```python
# Reviewer: Daniel Braunstain

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler

mnist = fetch_openml('mnist_784', version=1, as_frame=False)
X = mnist.data
y = mnist.target.astype(int)

# Plot the dots of one image
image = X[0].reshape(28, 28)  # Reshape the flat data into a 28x28 image
plt.imshow(image, cmap='gray', interpolation='none')
plt.title(f'Label: {y[0]}')
plt.show()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA(n_components=154)
X_pca = pca.fit_transform(X_scaled)
```

```python
kmeans = KMeans(n_clusters=10, random_state=42, n_init=10)
kmeans.fit(X)
clusters = kmeans.predict(X)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', s=10)
plt.colorbar()
plt.title("k-means Clustering on MNIST (PCA-reduced)")
plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score, silhouette_score

ari = adjusted_rand_score(y, clusters)
print(f"Adjusted Rand Index: {ari}")

silhouette_avg = silhouette_score(X, clusters)
print(f"Silhouette Score: {silhouette_avg}")
```

```python
# Get cluster centers in PCA space
centers = kmeans.cluster_centers_
# centers_original = pca.inverse_transform(centers)

# Plot cluster centers
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for i, ax in enumerate(axes.flat):
    ax.imshow(centers[i].reshape(28, 28), cmap='gray')
    ax.set_title(f"Cluster {i}")
    ax.axis('off')
plt.show()
```

```python
from scipy.stats import mode

# Map clusters to true labels
labels = np.zeros_like(clusters)
for i in range(10):
    mask = (clusters == i)
    labels[mask] = mode(y[mask])[0]  # Assign the most common label

accuracy = np.mean(labels == y)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

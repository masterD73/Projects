```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
human = pd.read_csv("human.csv")
```

```python
sns.displot(human.Salary, kde=True)
```

```python
proportion_martial = (human.groupby("MaritalDesc").count() / human.count())["Salary"]
proportion_martial = proportion_martial.sort_values(ascending=False)
sns.barplot(proportion_martial,color="red")
```

```python
perform_salary = human.groupby(["PerformanceScore"])["Salary"].median()
sns.barplot(perform_salary)
print(perform_salary)
```

```python
h = sns.PairGrid(human[["Absences", "PerformanceScore", "EmploymentStatus", "DaysLateLast30", "EmpSatisfaction", "Salary"]])
h.map(sns.barplot, color="red")
plt.show()
```

```python
g = sns.PairGrid(human[["Sex", "RaceDesc", "Salary", "PerformanceScore"]])
g.map(sns.scatterplot, color="red")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
human = pd.read_csv("human.csv")

sns.boxplot(x="Sex", y="Salary", data=human,color='blue')
sns.boxplot(x="RaceDesc", y="Salary", data=human, color='red')
plt.xticks(rotation=45)
race_sex = human.groupby(["Sex", "RaceDesc"])["Salary"].mean()

```

```python
martial_late = human.groupby(["MaritalDesc", "Sex"])["DaysLateLast30"].mean()
martial_late.sort_values(ascending=False)
martial_late.plot(kind="bar", color="red")
```

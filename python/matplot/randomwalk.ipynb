```python
import random
import pandas as pd
import seaborn as sns


def random_walk_1d(n):
    walk = [0]
    for i in range(1, n):
        x = random.choice((-1, 1))
        walk.append(walk[i - 1] + x)
    return walk


def random_walk_2d(n):
    x, y = 0, 0
    walk = [(x, y)]
    for i in range(1, n):
        (dx, dy) = random.choice([(0, 1), (0, -1), (1, 0), (-1, 0)])
        x += dx
        y += dy
        walk.append((x, y))

    return walk
```

```python
from matplotlib import pyplot as plt

# 1D
num_walks = 500
walks = [0] * num_walks
for i in range(num_walks):
    walks[i] = random_walk_1d(num_walks)
    sns.lineplot(walks[i])
plt.xlabel("Steps")
plt.ylabel("Distance")
plt.title("Random Walk 1D")
plt.show()
```

```python
walks = [[0]] * num_walks
for i in range(1, num_walks):
    walks[i] = random_walk_1d(num_walks)
sns.displot(walks, legend=False)
```

```python
# 2D
num_walks_2d = 1000
walks =[]
for i in range(0,num_walks_2d):
    walk = random_walk_2d(num_walks_2d)
    walks.append(walk)
```
